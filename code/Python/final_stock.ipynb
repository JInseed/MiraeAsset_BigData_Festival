```python
import pandas as pd
import numpy as np
```

```python
infor = pd.read_csv('C:/Users/Administrator/Desktop/미래에셋 공모전/data/전종목정보.csv')
year = pd.read_csv('C:/Users/Administrator/Desktop/미래에셋 공모전/data/연도별시가총액및주가상장주식수.csv')
table = pd.read_csv('C:/Users/Administrator/Desktop/미래에셋 공모전/data/재무제표_전처리.csv')

infor['Code'] = infor['Code'].apply(lambda x : str(x).zfill(6))
year['Code'] = year['Code'].apply(lambda x : str(x).zfill(6))

merge1 = pd.merge(infor,year, how='left', on = 'Code')
stock = pd.merge(table,merge1, how='left', left_on = ['기업','연도'], right_on = ['Name','연도'])
```

# PER, PBR, ROE 등 여러 파생변수 생성

```python
stock['유동비율'] = round(stock['유동자산'] / stock['유동부채'], 2)
stock['EPS'] = round(stock['당기순이익'] / stock['상장주식수'] ,2)
stock['PER'] = round(stock['시가총액'] / stock['당기순이익'], 2)
stock['PBR'] = round(stock['시가총액'] / stock['자본총계'], 2)
stock['ROE'] = round(stock['PBR'] / stock['PER'], 2)
stock['PCR'] = round(stock['시가총액'] / stock['영업현금흐름'], 2)
stock['POR'] = round(stock['시가총액'] / stock['영업이익'], 2)
stock['PSR'] = round(stock['시가총액'] / stock['매출액'], 2)
stock['GP/A'] = round(stock['매출총이익'] / stock['자산총계'], 2)

stock = stock.sort_values(['기업','연도'], ignore_index = True)
stock['전년도대비증감율'] = (stock['종가'] - stock['종가'].shift(1)) / stock['종가'].shift(1)
stock['일년뒤증감율'] = (stock['종가'].shift(-1) - stock['종가']) / stock['종가']
```

```python
stock.loc[:,['기업','연도','유동비율','EPS','ROE','PER','PBR','PCR','POR','PSR','GP/A','종가','일년뒤증감율','전년도대비증감율']].head(20)
```

```python
stock.to_csv('final_stock.csv', index=False, encoding='UTF-8-sig')
```
