```python
import pandas as pd
import numpy as np
import FinanceDataReader as fdr
```

```python
etf = fdr.StockListing('ETF/KR')
etf_sec = etf[etf['Name'].str.contains('TIGER 200')]
```

```python
etf_price = pd.DataFrame()
start_date = '2022-12-01'
end_date = '2023-06-30'

for ix, row in etf_sec.iterrows():
    code, name = row['Symbol'], row['Name']
    print(code, name)
    # 개별 종목의 가격 가져오기
    df = fdr.DataReader(code, '2022-11-29', '2023-07-03')

    # 가격 데이터의 종가(Close)를 컬럼으로 추가
    # 컬럼명은 종목명을 지정
    etf_price[name] = df['Close']
```

```python
etf_daily_profit = etf_price.pct_change(1).dropna()
etf_daily_profit
```
